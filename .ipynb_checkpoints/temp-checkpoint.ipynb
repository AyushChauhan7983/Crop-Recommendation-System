```python
grouped = crop.groupby(by='label').mean().reset_index()
grouped
```

```python
print(f'********************************')
for i in grouped.columns[1:]:
    print(f'Top 5 Most {i} requiring crops:')
    print(f'********************************')
    for j ,k in grouped.sort_values(by=i,ascending=False)[:5][['label',i]].values:
        print(f'{j} --> {k}')
    print(f'********************************')
```

```python
print(f'********************************')
for i in grouped.columns[1:]:
    print(f'Top 5 Least {i} requiring crops:')
    print(f'********************************')
    for j ,k in grouped.sort_values(by=i)[:5][['label',i]].values:
        print(f'{j} --> {k}')
    print(f'********************************')
```

```python
X = crop.drop(['crop_num','label'],axis=1)
y = crop['crop_num']
```

```python
X
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
acc = []
model =[]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisiontree = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)
decisiontree.fit(X_train,y_train)
predicted_values = decisiontree.predict(X_test)
x = metrics.accuracy_score(y_test,predicted_values)
acc.append(x)
model.append('Decision Tree')
print("DecisionTree's Accuracy is : ",x*100)
print(classification_report(y_test,predicted_values))
```
